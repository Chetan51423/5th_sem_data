```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
img = cv2.imread(r"C:\Users\cheta\Downloads\kaliya.png")
img =  cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
h, w = img.shape[:2]

x = int(input("Enter the x-axis shift; "))
y = int(input("Enter the y-axis shift; "))

t_matrix = np.float32([[1,0,x], [0,1,y]])
n_img = cv2.warpAffine(img, t_matrix, (w,h))

plt.imshow(img)
plt.show()
plt.imshow(n_img)
plt.show()
```
