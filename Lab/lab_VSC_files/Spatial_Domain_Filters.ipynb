High Pass Filter

```python
def high_pass(path):
    
    kernel = np.array([[-1, -1, -1],
                       [-1, 8, -1],
                       [-1, -1, -1]])

    
    # another kernal without 1/9
    out = cv2.filter2D(src=image, ddepth=-10, kernel=kernel)

    plt.imshow(image, 'gray')
    plt.title('Input Image')
    plt.show() 
    
    plt.imshow(out, 'gray')
    plt.title('High Pass Filter')
    plt.show()     
```
