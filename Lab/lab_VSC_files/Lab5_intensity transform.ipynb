Intensity level slicing without background

```python
image = cv2.imread(r"C:\Users\cheta\Downloads\sunset.jpeg")

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
x,y = image.shape[:2]

plt.imshow(gray,'gray')
plt.title('Original Image')
plt.show()
```

```python
z = np.zeros((x,y))
for i in range(0,x):
    for j in range(0,y):
        if(gray[i][j]>50 and gray[i][j]<150):
            z[i][j]=255
        else:
            z[i][j]=0
plt.imshow(z, 'gray')
plt.title('Sliced Original Image')
plt.show()
```

Intensity level slicing with background

```python
image = cv2.imread(r"C:\Users\cheta\Downloads\sunset.jpeg")

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
x,y = image.shape[:2]

plt.imshow(gray,'gray')
plt.title('Original Image')
plt.show()

z = np.zeros((x,y))
for i in range(0,x):
    for j in range(0,y):
        if(gray[i][j]>50 and gray[i][j]<150):
            z[i][j]=255
        else:
            z[i][j]=gray[i][j]

plt.imshow(z, 'gray')
plt.title('Sliced with background')
plt.show()
```

Log transformation

```python
image = cv2.imread(r"C:\Users\cheta\Downloads\sunset.jpeg")

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
x,y = image.shape[:2]

plt.imshow(gray,'gray')
plt.title('Original Image')
plt.show()
```

```python
c = 255/(np.log(1+np.max(gray)))

log = c*np.log(1+gray)
log = np.array(log,dtype=np.uint8)

plt.imshow(log, 'gray')
plt.show()
```
