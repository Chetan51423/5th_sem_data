```python
#importing openCV C++ library who provide wide range of functions
import cv2

#importing python library for numerical computing 
import numpy as np

#inlcluding python library for data visualization
from matplotlib import pyplot as plt

#imposrting specific functions and object from numpy library
from numpy import dot, exp, mgrid, pi, ravel, square, uint8, zeros
```

```python
#reading image form user
img = cv2.imread(r"C:\Users\cheta\Downloads\sunset.jpg",0)

#doing fourier transform of original image
f = np.fft.fft2(img)

# Shift the zero frequency component to the center of the spectrum
fshift = np.fft.fftshift(f)

#finding magnitude spectrum
magnitude_spectrum = 20*np.log(np.abs(fshift))


#findidng phase spectrum
phase_spectrum = np.angle(fshift)

plt.subplot(131),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])

plt.subplot(132),plt.imshow(magnitude_spectrum, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])

plt.subplot(133),plt.imshow(phase_spectrum, cmap = 'gray')
plt.title('Phase Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()
```

Perform high pass fitltering and reconstruct the image by removing the 100 x 100 region from the input image

```python
import cv2
import numpy as np
```

```python
# Load the input image in grayscale
img = cv2.imread(r"C:\Users\cheta\Downloads\sunset.jpeg", 0)

# Perform Fourier transform
f = np.fft.fft2(img)

# Shift the zero frequency component to the center of the spectrum
fshift = np.fft.fftshift(f)

# Create a high pass filter mask
rows, cols = img.shape
crow, ccol = rows // 2, cols // 2
mask = np.ones((rows, cols), np.uint8)
mask[crow-50:crow+50, ccol-50:ccol+50] = 0

# Apply the high pass filter mask
fshift_filtered = fshift * mask

# Shift the zero frequency component back to the top left corner
f_filtered = np.fft.ifftshift(fshift_filtered)

# Perform inverse Fourier transform
img_filtered = np.fft.ifft2(f_filtered)

# Normalize the filtered image to bring it back to its original intensity range
img_filtered = np.abs(img_filtered)
img_filtered = cv2.normalize(img_filtered, None, 0, 255, cv2.NORM_MINMAX, cv2.CV_8U)

# Display the input and filtered images
cv2.imshow('Input Image', img)
#cv2.imshow('Fourier transfor image', f)
#cv2.imshow('shift to center of the spectrum', fshift)
#cv2.imshow('Img after applying high pass filter', fshift_filtered)
#cv2.imshow('Image shifter to zero again', f_filtered)
cv2.imshow('Inverse fourier transform', img_filtered)
#cv2.imshow('Filtered Image', img_filtered)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
